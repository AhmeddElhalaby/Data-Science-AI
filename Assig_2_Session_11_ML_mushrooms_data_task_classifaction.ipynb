```python
le = LabelEncoder()
df = df.apply(le.fit_transform)
```

```python
df.drop(columns=["veil-type"] , inplace=True)
```

```python
df.corr()["class"]
```

```python
x = df.drop(columns=['class'])
y = df["class"]
x_train , x_test , y_train , y_test = train_test_split(x , y , train_size=.7 , random_state= 42)
```

```python
RF = RandomForestClassifier()
param_grid = {
    "max_depth":[500,1000,2000] ,
    #"min_samples_split":[2,40,80] ,
    "max_features": [10,22],
    "n_estimators":[10,50,100]
}                                        ### no.=3*3*2=18

grid_search = GridSearchCV(estimator = RF , param_grid = param_grid , cv = 5 ,
                           scoring = "accuracy" , n_jobs = -1 , verbose = 1 )   ### cv for K-Fold   ### no.=18*5=90
                                                                                    ## scoring for determinig best parameter
                                                                                    ## n_jobs for processor to be multi task processor
                                                                                    ## verbose 1 for showing some hidden data like no. of trials

grid_search.fit(x_train , y_train)
best_model = grid_search.best_estimator_
y_pred = best_model.predict(x_test)
acc = accuracy_score(y_test , y_pred)
# print(grid_search.best_params_)
# print(grid_search.best_estimator_)
# print(acc)
```

```python
print(grid_search.best_params_)
print(grid_search.best_estimator_)
print(acc)
```

```python
from sklearn.ensemble import StackingClassifier

# define base models  as   list of tupples(,)
base_models =[
    ('lr' , LogisticRegression(random_state= 42 , max_iter=1000 )),
    ('dt' , DecisionTreeClassifier(random_state= 42 )),
    ('rf' , RandomForestClassifier(n_estimators=10 , random_state=42)),
    ('knn', KNeighborsClassifier(n_neighbors=5 ))
] 
# define meta_model
meta_model = LogisticRegression(random_state=42 , max_iter=1000)

# stacking classifier 
stacking_clf = StackingClassifier(estimators=base_models , final_estimator= meta_model , cv = 5 )   ## cv sum of base models and meta model

# train stacking_clf
stacking_clf.fit(x_train , y_train)

y_pred = stacking_clf.predict(x_test)
acc = accuracy_score(y_test , y_pred)
print("StackingClassifier accuracy:" , acc)
```
